```python
from datetime import date
def tstp(t):
    return date.fromtimestamp(t)
df['film_date2'] = df['film_date'].apply(tstp)
df[['film_date', 'film_date2']].iloc[0:10]
```

```python
def flow(i, **kwargs):
    tag_list = eval(i)
    if kwargs['targets'] in tag_list:
        return True
    else:
        return False
s = input('輸入篩選的標籤:')

target_df = df[ df['tags'].apply(flow, targets=s) ]
target_df.to_csv('filter.csv', encoding='utf-8', index=False)
target_df
```
